```python
import matplotlib.pyplot as plt 
import numpy as np


x = np.linspace(0,2 * np.pi, 200)
y = np.sin(x)


fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["font.sans-serif"]=["SimHei"] #设置字体
plt.rcParams["axes.unicode_minus"]=False #正常显示负号
year =[2017,2018,2019,2020]
people=[20,40,60,70]
#生成图表
plt.bar (year,people)
plt.xlabel('年份')
plt.ylabel('人口')
plt.title('人口增长') 
#设置纵坐标刻度
plt.yticks([0,20,40,60,80])
#填充颜色
for i in range(len(year)):
    if people[i] < 20:
        plt.fill_between(year[i], people[i], 20, color='yellow')
# 显示图像
plt.show()
```
