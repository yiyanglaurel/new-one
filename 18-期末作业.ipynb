```python
# 生活满意度与信息判断能力的相关性、
def 制作交叉表(数据表, 自变量, 因变量):
    # 从数据表中提取出自变量和因变量的数据
    自变量数据 = 数据表[自变量]
    因变量数据 = 数据表[因变量]

    # 制作交叉表
    return pd.crosstab(自变量数据, 因变量数据, normalize='columns', margins=True)
```
