```python
import matplotlib.pyplot as plt
import pandas as pd


def plot_variable_pie(data_table, variable):
    # 准备绘图数据
    data = data_table[variable].value_counts()
    # 创建饼图
    fig1, ax1 = plt.subplots()
    ax1.pie(data, labels=data.index, autopct='%1.1f%%')
    # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.axis('equal')

    plt.show()
```

```python
import mytools
df = pd.read_csv(R'data\类别变量练习数据.csv', encoding='GBK')
plot_variable_pie(df, '父亲职业重新编码')
```

```python
def 制作交叉表(数据表, 自变量, 因变量):
    # 从数据表中提取出自变量和因变量的数据
    自变量数据 = 数据表[自变量]
    因变量数据 = 数据表[因变量]

    # 制作交叉表
    return pd.crosstab(自变量数据, 因变量数据, normalize='columns', margins=True)
```
